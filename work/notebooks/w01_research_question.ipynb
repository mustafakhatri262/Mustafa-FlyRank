```python
from pathlib import Path
import pandas as pd
import os

github_repo_url = 'https://github.com/mustafakhatri262/Mustafa-FlyRank.git'

repo_dir_name = None
if github_repo_url != 'YOUR_GITHUB_REPO_URL':
    repo_dir_name = Path(github_repo_url).stem
    if not Path(repo_dir_name).exists():
        print(f"Cloning '{github_repo_url}' into '{repo_dir_name}'...")
        !git clone {github_repo_url}
        print(f"Repository '{repo_dir_name}' cloned successfully.")
    else:
        print(f"Repository '{repo_dir_name}' already exists. Skipping clone.")
else:
    print("Please set 'github_repo_url' to clone your repository.")

repo_root = None
data_path = None
csv_filename = "content_refresh_anonymized.csv"
relative_data_path = Path("data") / "raw" / csv_filename

# 1. Try to find the file within the potentially cloned repository first
if repo_dir_name and Path(repo_dir_name).exists():
    candidate_in_cloned_repo = Path(repo_dir_name) / relative_data_path
    if candidate_in_cloned_repo.exists():
        repo_root = Path(repo_dir_name)
        data_path = candidate_in_cloned_repo

# 2. Fallback to original search logic if not found in cloned repo
if repo_root is None:
    for base in [Path.cwd(), *Path.cwd().parents]:
        candidate = base / relative_data_path
        if candidate.exists():
            repo_root = base
            data_path = candidate
            break

if repo_root is None:
    raise FileNotFoundError(f"Could not find {relative_data_path} in the cloned repository or local paths.")

df = pd.read_csv(data_path)

rows = len(df)
clients = df["client_id"].nunique()
declining = (df["trend_direction"].str.lower() == "down").sum()
declining_rate = declining / rows

print(f"Loaded: {data_path}")
print(f"Rows: {rows:,}")
print(f"Clients: {clients}")
print(f"Declining pages: {declining:,} ({declining_rate:.1%})")
```
